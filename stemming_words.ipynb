```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import pickle
import collections

ps = PorterStemmer()

sentence = "gaming, the gamers play games"
words = word_tokenize(sentence)

for word in words:
    print(word + ":" + ps.stem(word))
```

```python
with open('topics.pk', 'rb') as handle:
    topics_list = pickle.load(handle)
```

```python
counts = {}
doc.keys()

for doc in topics_list:
    print(doc["topic_name"])
    words = word_tokenize(doc["topic_name"])
    for word in words:
        word = word.lower()
        if ps.stem(word) in counts:
            counts[ps.stem(word)] += 1
        else:
            counts[ps.stem(word)] = 1
        
# print(doc["discussion_material"])
# sorted(counts, key=counts.get)
sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)
```

```python
stop_words = ['?',
              'explain',
                'what',
                'is',
                'the',
                'a',
                'of',
                'in',
                'are',
                'an',
                '.',
                'it',
                'for',
                'thi',
                'with',
                'be',
                '–',
                'how',
                'to',
                'when',
                'and',
                'you',
                'can',
                ',',
                'from',
                'better',
                'do',
                'that',
                'by', 
                'on',
                'as',
                'or',
                ')',
                '('
              ]
```

```python
for i in stop_words:
    counts[i]=0

sorted_counts = sorted(counts.items(), key=lambda x: x[1], reverse=True)
```

```python
sorted_counts
```

### Discussion Material

```python
counts_disc = {}

for doc in topics_list:
#     print(doc["discussion_material"])
    words = word_tokenize(doc["discussion_material"])
    for word in words:
        word= word.lower()
        if ps.stem(word) in counts_disc:
            counts_disc[ps.stem(word)] += 1
        else:
            counts_disc[ps.stem(word)] = 1
```

```python
for i in stop_words:
    counts_disc[i]=0

sorted_counts_disc = sorted(counts_disc.items(), key=lambda x: x[1], reverse=True)
```

```python
sorted_counts_disc
```
